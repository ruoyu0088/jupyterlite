````python
def writable_array(s):
    info = s.to_numpy().__array_interface__
    info['data'] = info['data'][0], False
    s.__array_interface__ = info
    return np.array(s, copy=False)
````

````python
s = df.get_column('A')
arr = writable_array(s)
arr[:2] = 100, 200
````
